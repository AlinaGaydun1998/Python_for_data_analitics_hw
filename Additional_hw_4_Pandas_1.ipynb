6. Трансформуйте колонки 'Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків' у тип даних float. УВАГА! Перед трансформацією подивіться, які дані містять ці колонки. Аби успішно трансформувати тип даних, треба спочатку привести рядки до того виду, який буде підходити, аби потім рядок зробити типом float.

```python
# перевіримо які дані містить наш датасет
print(f'Колонка "Ваш вік містить значення": {df["Ваш вік"].unique()}'),
print(f'Колонка "Зарплата у $$$ за місяць" містить значення: {df["Зарплата у $$$ за місяць, лише ставка після сплати податків"].unique()}')
```

```python
# очистимо дані від пропущених значень та зайвих символів

def clean_data(df, args):
    for col in args:
        df[col] = df[col].replace(',', '.')
        df = df.dropna(subset= [col])
    return df

# перетворення даних до типу float

def transform_to_float(df, args):
    for col in args:
        df[col] = pd.to_numeric(df[col], errors = 'coerce')
    return df  
```

```python
# застосуємо зміни

df = clean_data(df, ['Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків'])
df = transform_to_float(df, ['Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків'])

# переглянемо нові дані

print(f'Колонка "Ваш вік містить значення": {df["Ваш вік"].unique()}'),
print(f'Колонка "Зарплата у $$$ за місяць" містить значення: {df["Зарплата у $$$ за місяць, лише ставка після сплати податків"].unique()}') 
```

```python
# варіант перетворення через пандас

df['DateTime'] = pd.to_datetime(df['Timestamp'])
df['DateTime'].iloc[:5], df['DateTime'].iloc[-5:]
```

```python
# Перетворення через datetime

from datetime import datetime

def convert_to_data(timestamp):
    try:
        # Перевіряємо, чи є дата в форматі 'MM.DD.YYYY HH:MM:SS'
        if '.' in timestamp.split(' ')[0] and len(timestamp.split(' ')[0].split('.')) == 3:
            # Міняємо місцями місяць і день
            month, day, year = timestamp.split(' ')[0].split('.')
            timestamp = f"{day}.{month}.{year} {timestamp.split(' ')[1]}"  # Міняємо місцями день і місяць
        
        # Перевіряємо на формат 'DD.MM.YYYY HH:MM:SS'
        if len(timestamp.split(' ')[1]) == 8:  # Це означає, що є секунди
            return datetime.strptime(timestamp, '%d.%m.%Y %H:%M:%S')
        else:  # Це формат без секунд, 'DD.MM.YYYY HH:MM'
            return datetime.strptime(timestamp, '%d.%m.%Y %H:%M')
        
    except ValueError:
            # Якщо не вдалося перетворити, повертаємо NaT
        return pd.NaT
```

```python
df['Timestamp'] = df['Timestamp'].apply(convert_to_data)
df['Timestamp']
```

8. Для колонок типу даних object (які такими лишились після всіх наших трансформацій) виведіть кількість унікальних значень.

```python
# Фільтруємо тільки колонки з типом 'object'
filtered = df.select_dtypes(include='object')

# Рахуємо унікальні значення для кожної колонки
for col in filtered.columns:
    print(f"Column: {col}")
    print(filtered[col].value_counts())
    print("\n")
```

```python
# Ініціалізуємо список для зберігання унікальних значень об'єктних колонок
unique_objects = []

# Перебираємо всі колонки у DataFrame
for col in df.columns:
    # Перевіряємо, чи тип даних колонки - об'єкт (строковий)
    if df[col].dtype == 'object':
        # Рахуємо унікальні значення і зберігаємо їх як словник
        value_counts = df[col].value_counts().to_dict()
        # Додаємо колонку та її унікальні значення до списку
        unique_objects.append((col, value_counts))

# Створюємо DataFrame для зручного відображення результатів
unique_objects_df = pd.DataFrame(unique_objects, columns=['Column Name', 'Unique Values'])

# Налаштування для відображення повного тексту у колонках DataFrame
pd.set_option('display.max_colwidth', None)

# Виводимо DataFrame
display(unique_objects_df)
```

12. Зробіть зводну таблицю, де стовпці - це 'Ваш тайтл', рядки - 'Знання англійської мови', а значення - медіанна ЗП. Які висновки можна зробити?
Тут ще для зручності можна зробити градієнтне забарвлення і відформатувати значення для легшого перегляду.

```python
# Створюємо зведену таблицю (pivot table)
table = pd.pivot_table(
    df, 
    index='Знання англійської мови',  
    columns='Ваш тайтл',             
    values='Зарплата у $$$ за місяць, лише ставка після сплати податків',  
    aggfunc='median'                 
)

# Форматуємо значення у зведеній таблиці та додаємо градієнтне забарвлення
styled_pivot_table = (
    table.style
    .format("{:.2f}")               
    .background_gradient(cmap="RdYlBu")  
)

# Відображаємо стилізовану зведену таблицю
styled_pivot_table
```

```python
top_regions_frequency = (
    df['В якій області ви зараз живете?']
    .value_counts()
    .head(5)
    .index
)

# Фільтруємо дані, залишаючи лише записи з цих регіонів
filtered_df = df[df['В якій області ви зараз живете?'].isin(top_regions_frequency)]

# Обчислюємо описову статистику зарплат для кожного регіону
result = filtered_df.groupby('В якій області ви зараз живете?')[
    'Зарплата у $$$ за місяць, лише ставка після сплати податків'
].describe()

# Сортуємо результат за кількістю записів (count) у спадному порядку
sorted_result = result.sort_values(by='count', ascending=False)

# Відображаємо відсортовану таблицю з результатами
display(sorted_result)
```

14. Як зростає зарплата з роками? Для цього нам треба відформатувати колонку "Загальний стаж роботи за спеціальністю" аналогічно до того, як ми це робили для колонки 'Знання англійської мови' і потім вивести медіанні значення для кількості років стажу.

```python
# Встановлення порядку для 'Загальний стаж роботи за спеціальністю'
order = [
    "Менше як 3 місяці", "3 місяці", "Пів року", "1 рік", "1,5 року", "2 роки",
    "3 роки", "4 роки", "5 років", "6 років", "7 років", "8 років", "9 років",
    "10 років", "11 років", "12 років", "13 років", "14 років", "15 і більше років"
]

# Перетворюємо на категоріальний тип даних колонку 'Загальний стаж роботи за спеціальністю'
df['Загальний стаж роботи за спеціальністю'] = pd.Categorical(
    df['Загальний стаж роботи за спеціальністю'], categories=order, ordered=True
)

# Обчислюємо медіанну зарплату залежно від загального стажу роботи за спеціальністю
salary_increase = df.groupby('Загальний стаж роботи за спеціальністю')[
    'Зарплата у $$$ за місяць, лише ставка після сплати податків'
].median()

# Відображаємо результати
display(salary_increase)
```

Зарплата відповідно до стажу зростає пропорційно: чим вищій стаж - тим вища заробітна плата. Винятком в нашій вибірці є стаж 8 років де медіанна з/п становить 4700 що є вищім ніж при стажі 9-11 років. Можливо це пов'язано з тим, що співробітники зі стажем 8 років, ймовірно, займають посади з вищими обов’язками чи рівнем кваліфікації, отримуючи вищу оплату. Також це може свідчити про обмежену вибірку для стажів 9-12 років або нерівномірний розподіл посад у цих категоріях. Додатковий аналіз даних за посадами та галузями може допомогти уточнити причину цієї аномалії.

Далі тут можна зробити багато аналогічних агрегацій, проаналізувати, в який час люди заповнювали форму, хто більше отримує, Business Analyst, чи Data Analyst. Я вас заохочувати не стримувати свій інтерес і покапати ці дані далі :)

```python
# Створюємо зведену таблицю для аналізу медіанної зарплати залежно від стажу та посади
table_years_title_salary = pd.pivot_table(
    df,
    index='Загальний стаж роботи за спеціальністю',  
    columns='Оберіть вашу посаду',  
    values='Зарплата у $$$ за місяць, лише ставка після сплати податків',  
    aggfunc='median'  
)

# Додаємо стиль: форматування значень до 2-х знаків після коми та градієнтне забарвлення
styled_table_years = table_years_title_salary.style.format('{:.2f}').background_gradient(cmap='coolwarm')

# Відображаємо стилізовану таблицю
styled_table_years
```

Згідно зведеної таблиці вище, цікаво прослідковується що в спеціалізації DBA стаж роботи від 2 років і вище, що може свідчити про доволі такі складні задачі в сфері адміністрування базами і новачків дуже мало в цій сфері. А в Customer Success навпаки більше 10-11 років ніхто не затримується і можливо після такого досвіду вже змінює спеціалізацію на Project/Product/Program Manager що вже перспективніше в розрізі розвитку зарплати та карьєрних можливостей.
Загалом розкид зростання зарплат до стажу роботи є не зовсім однорідний, що може сказати що все таки навички і вміння брати відповідальність і вирішувати складні задачі цінується більше ніж просто цифра у стажі роботи. 

Оскільки мене цікавить напрямок розвитку Data Analyst, Data Science, Machine Learning, Big Data, то подальший аналіз проведу для цих груп напрямку спеціалізації. 

```python
# Створюємо зведену таблицю для аналізу гендерного розподілення за посадами
pivot_table_analytics = pd.pivot_table(
    df, 
    index= 'Ваша стать',
    columns = 'Ваша посада4', 
    aggfunc='size',
)
pivot_table_science = pd.pivot_table(
    df, 
    index= 'Ваша стать',
    columns = 'Ваша посада5',
    aggfunc='size',
)

# Об'єднання двох півот-таблиць
pivot_table_combined = pd.concat([pivot_table_analytics, pivot_table_science], axis=1)

# Виведення результату
display(pivot_table_combined)
```

```python
# Створюємо зведену таблицю для аналізу використання програмування на позиціях аналітиків та дата саєнтистів
pivot_analytics_programming = pd.pivot_table(
    df, 
    index= 'Чи використовуєте ви у своїй роботі мови програмування (одну чи декілька)?',
    columns = 'Ваша посада4', 
    aggfunc='size',
)
pivot_science_programming = pd.pivot_table(
    df, 
    index= 'Чи використовуєте ви у своїй роботі мови програмування (одну чи декілька)?',
    columns = 'Ваша посада5',
    aggfunc='size',
)

# Об'єднання двох півот-таблиць
pivot_programming_combined = pd.concat([pivot_analytics_programming, pivot_science_programming], axis=1)

# Виведення результату
display(pivot_programming_combined)
```
